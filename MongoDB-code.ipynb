QUERIES

QUERY 1 - How many articles mention Apple? ---proposed query 


```python
#Count number of subdocuments that include the word "Apple" 

collection.count_documents({})
```

QUERY 2 - How many articles are about Apple? 

```python

#Count number of subdocuments that include the word "Apple" on the description

query2 = {
    "content": {
        "$regex": 'Apple',
        "$options" :'i' # case-insensitive
    }
}

results_news2 = collection.find(query2)
print(collection.count_documents(query2))
```

QUERY 3 - How many articles are about Apple? 

```python

#Count number of subdocuments that include the word "Apple" on the title
query3 = {
    "title": {
        "$regex": 'Apple',
        "$options" :'i' # case-insensitive
    }
}

results_news3 = collection.find(query3)
print(collection.count_documents(query3))
```
